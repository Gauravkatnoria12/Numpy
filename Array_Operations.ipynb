**Stacking Array**

```python
A = np.array([1,2,3,4,5])
B = np.array([6,7,8,9,10])
np.vstack((A, B))   # Vertical stack
```

```python
np.hstack((A, B))   # Horizontal Stack
```

```python
np.column_stack((A, B))     # 2D column stack
```
